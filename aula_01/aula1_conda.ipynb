# Leitura de imagens

```python
grid = cv2.imread("img9x9.png")
```

```python
grid_rgb = cv2.cvtColor(grid, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(grid_rgb,interpolation="none")
```
